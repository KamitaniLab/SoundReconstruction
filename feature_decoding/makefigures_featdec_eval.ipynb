```python
'''Make figures of feature decoding results.'''


import os
import yaml

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Functions #####################################################

def weird_form_to_long(df, target_col, identify_cols=[]):
    df_result = pd.DataFrame()
    for i, row in df.iterrows():
        tmp = {}
        for col in identify_cols:
            tmp[col] = row[col]
        tmp[target_col] = row[target_col]
        df_result = pd.concat([df_result, pd.DataFrame(tmp)])
    return df_result
```

```python
# Settings #######################################################
conf_file = "config/vggsound_fmriprep_rep4_500voxel_vggishish_allunits_fastl2lir_alpha100.yaml"
output_dir = '../data/figures'

with open(conf_file, 'r') as f:
    conf = yaml.safe_load(f)
analysis_name = conf['analysis name']

decoding_accuracy_file = os.path.join(
    conf['decoded feature dir'],
    analysis_name,
    'decoded_features',
    conf['network'],
    'accuracy.pkl.gz'
)
if 'test fmri' in conf:
    subjects = list(conf['test fmri'].keys())
else:
    subjects = list(conf['fmri'].keys())
rois = list(conf['rois'].keys())
layers = conf['layers']

print("Decoding accuracy filepath:", decoding_accuracy_file)
print("Subjects: {}".format(subjects))
print("ROIs:     {}".format(rois))
print("Features: {}".format(layers))
perf_df = pd.read_pickle(decoding_accuracy_file)
display(perf_df)

```

```python
# Prepare the data for plotting #################################
a_df = perf_df.query("layer == 'conv5_3'")
corr_df = a_df[["subject", "roi", "profile correlation"]]
corr_df = weird_form_to_long(corr_df, "profile correlation", ["subject", "roi"])

ident_df = a_df[["subject", "roi", "identification accuracy"]]
ident_df = weird_form_to_long(ident_df, "identification accuracy", ["subject", "roi"])
```

```python
# Fig.3D Correlation coefficient #################################
sns.set_style("whitegrid", {'grid.linestyle': '--'})

fig, ax = plt.subplots(1, 1, figsize=(12, 4))
sns.barplot(ax=ax, 
            data=corr_df, 
            x="roi", 
            y="profile correlation",  
            hue="subject", 
            ci=None,
            palette=["mediumblue", "cornflowerblue", "limegreen", "yellow", "orangered"])
ax.set_ylim([0, 0.7])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
save_filename = os.path.join(output_dir, 'Fig3D.png')
fig.savefig(save_filename, dpi=300, bbox_inches='tight', pad_inches=0.05)
print('Saved {}'.format(save_filename))
```

```python
# Fig.3E Identification accuracy #################################
fig, ax = plt.subplots(1, 1, figsize=(12, 4))
sns.barplot(ax=ax, 
            data=ident_df, 
            x="roi", 
            y="identification accuracy",  
            hue="subject", 
            palette=["mediumblue", "cornflowerblue", "limegreen", "yellow", "orangered"])
ax.set_ylim([0.4, 1.0])
ax.axhline(0.5, color='k', linestyle='-')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
save_filename = os.path.join(output_dir, 'Fig3E.png')
fig.savefig(save_filename, dpi=300, bbox_inches='tight', pad_inches=0.05)
print('Saved {}'.format(save_filename))

```
